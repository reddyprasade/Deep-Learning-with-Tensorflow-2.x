### Loading Multiple image From Internet

```python
import numpy as np
import os
import PIL
import PIL.Image
import tensorflow as tf
import tensorflow_datasets as tfds
import pathlib
```

```python
dataSet_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
dataSet_url
```

```python
data_dir = tf.keras.utils.get_file(origin=dataSet_url,
                                  fname = "flower_photos",
                                   untar=True)
```

```python
data_dir = pathlib.Path(data_dir)
data_dir
```

```python
image_count = len(list(data_dir.glob('*/*.jpg')))
image_count
```

```python
tulips = list(data_dir.glob('tulips/*'))
PIL.Image.open(str(tulips[0]))
```

```python
daisy = list(data_dir.glob('daisy/*'))
PIL.Image.open(str(daisy[0]))
```

```python
roses = list(data_dir.glob('roses/*'))
PIL.Image.open(str(roses[0]))
```

i will Create a Data with Height and width of image

```python
batch_size = 32
img_height = 250
img_width  = 250
```

It's good practice to use a validation split when developing your model. We will use 80% of the images for training, and 20% for validation.

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
                                                            data_dir,
                                                            validation_split =0.2,
                                                            subset="training",
                                                            seed=123,
                                                            image_size = (img_height,img_width),
                                                            batch_size=batch_size)                                                            
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
val_ds.class_names
```

```python
class_names =train_ds.class_names
class_names
```

```python
plt.figure(figsize=(10,10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3,3,i+1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
for images_batch, labels_batch in train_ds:
    print(images_batch.shape)
    print(labels_batch.shape)
    break
```

### Data Normilization in TensorFlow 

* Hear we now that We have (RGB Chanel) and this Values From 0 to 255 we t=have bring this data into Standard values [0, 1] by using Resize

```python
normalization_layer  = tf.keras.layers.experimental.preprocessing.Rescaling(1./255)
```

```python
normalized_ds  = train_ds.map(lambda x,y:(normalization_layer(x),y))
images_batch, labels_batch = next(iter(normalized_ds))
```

```python
first_image = images_batch[0]
```

```python
np.min(first_image), np.max(first_image)
```

```python
num_classes = 5
```

```python
model = tf.keras.models.Sequential([
                                    tf.keras.layers.experimental.preprocessing.Rescaling(1./255),
                                    tf.keras.layers.Conv2D(16,3, padding='same',activation ='relu'),  # Con
                                    tf.keras.layers.Dense(num_classes)
                                   ])
model.compile(optimizer='sgd',loss='categorical_crossentropy')
```
